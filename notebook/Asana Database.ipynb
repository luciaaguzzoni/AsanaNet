```python
kaggle_path_1 = "../data/archive/final_asan1_1.csv"
df1 = pd.read_csv(kaggle_path_1)
df1.drop("AID", axis=1, inplace=True)
df1["AName"] = df1["AName"].apply(lambda x : x.lower())
df1.sort_values(by=["AName"],inplace=True,ignore_index=True)

# Drop rows containing specified phrases in the 'AName' column
phrases_to_drop = ['blinking', 'front and sideways viewing', 'nearing and distant viewing',
                   'rotational viewing', 'sideways viewing', 'up and down viewing', 'prayanama', 'yoni mudra']
df1 = df1[~df1['AName'].str.contains('|'.join(phrases_to_drop))]
```

```python
# Use regular expression to extract values within parentheses
df1['AName'] = df1['AName'].str.replace('\n',' ')
df1['ENG'] = df1['AName'].str.extract(r'\((.*?)\)')
#df1['ENG'] = df1["AName"].str.extract(r'\((.*$)')
df1['AName'] = df1['AName'].replace(to_replace=r'\((.*?)\)', value='', regex=True)
#df1['AName'] = df1['AName'].replace(to_replace=r'\((.*$)', value='', regex=True)
df1['AName'] = df1['AName'].str.strip()
```

```python
df1.loc[df1["AName"] == 'vrischikasana (scorpion pose',"ENG"] = 'scorpion pose'
df1.loc[df1["AName"] == 'vrischikasana (scorpion pose',"AName"] = 'vrischikasana'

df1.loc[df1["AName"] ==  'sukhasana (easy pose',"ENG"] = 'easy pose'
df1.loc[df1["AName"] == 'sukhasana (easy pose',"AName"] = 'sukhasana'

df1.loc[df1["AName"] ==  "hanumanasana (hanuman's pose","ENG"] = "hanuman's pose"
df1.loc[df1["AName"] ==  "hanumanasana (hanuman's pose","AName"] = "hanumanasana"
```
