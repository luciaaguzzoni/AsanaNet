8. **DEFINE** (ENCAPSULATE) into a FUNCTION
    - Function name: profit_per_customer
    - Function args: name, string. the name of the customer to filter by
        - to do this, you will have to change the line where you filter to change the literal "Roy Franz√∂sisch" for name
        - also change it in the formatted print
    - Function should do steps: 1-8
    - include the line: `df_imported = pd.read_csv("customer_example.csv")`
    - Function return: it should return the `df_imported`

```python
def profit_per_customer(name):
    dataset = pd.read_csv("../Documents/Ironhack/bootcamp/lectures/datasets/superstore.csv")
    df = pd.DataFrame(dataset)
    df.columns = [i.lower().replace(" ","_") for i in df.columns]
    df.drop(columns="row_id",inplace=True)
    client_df = df[df["customer_name"]==name]
    client_profit = sum(list(client_df["profit"]))
    print(f"{name} amounted a total profit of {client_profit} $")
    client_df.to_csv("./customer_example.csv")
    #client_df.to_csv("./customer_example.csv",index=False)
    df_imported = pd.read_csv("customer_example.csv")
    return df_imported
```
